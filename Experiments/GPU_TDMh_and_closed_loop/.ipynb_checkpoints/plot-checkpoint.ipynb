```python
import numpy as np
import matplotlib.pyplot as plt

# Helper function to read and process data
def read_data(file_path):
    with open(file_path, 'r') as f:
        return [float(line.strip()) for line in f.readlines()]

# Load all required data
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial']
})
open_s1 = read_data('tdmalogs1/rtr1.txt')
open_s2 = read_data('tdmalogs1/rtr2.txt')
open_p1 = read_data('tdmalogs1/misses1.txt')
open_p2 = read_data('tdmalogs1/misses2.txt')

closed_s11 = read_data('logs/s11.txt')
closed_s22 = read_data('logs/s22.txt')
closed_cpt1 = read_data('logs/deadlinemisst1.txt')
closed_cpt2 = read_data('logs/deadlinemisst2.txt')

# Setpoint lines
setpoint_value = 0.90
setpoint_value1 = 0.90 
setpoint_open = [setpoint_value] * len(open_s1)
setpoint_open1 = [setpoint_value1] * len(open_s1)
setpoint_closed = [setpoint_value] * len(closed_s11)

# Define index slices
idx_rtr = slice(1, 50)
idx_period = slice(1, 50)

# Setup plot
fig, axs = plt.subplots(1, 4, figsize=(11, 3), sharex=False)

# --- Subplot 0: Open-loop RTR ---
axs[0].plot(open_s1[idx_rtr], '-', linewidth=2, label='mm1')
axs[0].plot(open_s2[idx_rtr], '-.', linewidth=2, label='mm2')
axs[0].plot(np.array(setpoint_open)[idx_rtr], '--', linewidth=2, label='Set Point 1')
axs[0].plot(np.array(setpoint_open1)[idx_rtr], '--', linewidth=2, label='Set Point 2')
axs[0].set_ylabel('RTR', fontsize=13)
axs[0].set_xlabel('Control Period\n\n(a) GPU-TDMh', fontsize=12)
axs[0].set_ylim([0, 2])
axs[0].spines['top'].set_visible(False)

# --- Subplot 1: Open-loop Period ---
axs[1].plot(open_p1[idx_rtr], '-', linewidth=2)
axs[1].plot(open_p2[idx_rtr], '-.', linewidth=2)
axs[1].set_ylabel('Deadline misses (%)', fontsize=14)
axs[1].set_xlabel('Control Period\n\n(b) GPU-TDMh', fontsize=12)
#axs[1].set_ylim([0, 0.5])
axs[1].spines['top'].set_visible(False)

# --- Subplot 2: Closed-loop RTR ---
axs[2].plot(closed_s11[idx_period], '-', linewidth=2)
axs[2].plot(closed_s22[idx_period], '-.', linewidth=2)
axs[2].plot(np.array(setpoint_closed)[idx_period], '--', linewidth=2)
axs[2].plot(np.array(setpoint_open1)[idx_rtr], '--', linewidth=2)
axs[2].set_ylabel('RTR', fontsize=13)
axs[2].set_xlabel('Control Period\n\n(c) Closed-loop', fontsize=12)
axs[2].set_ylim([0, 2])
axs[2].spines['top'].set_visible(False)

# --- Subplot 3: Closed-loop Period ---
axs[3].plot(closed_cpt1[idx_period], '-', linewidth=2)
axs[3].plot(closed_cpt2[idx_period], '-.', linewidth=2)
axs[3].set_ylabel('Deadline misses (%)', fontsize=14)
axs[3].set_xlabel('Control Period\n\n(d) Closed-loop', fontsize=12)
#axs[3].set_ylim([0, 0.5])
axs[3].spines['top'].set_visible(False)

# Shared formatting
for ax in axs:
    ax.tick_params(axis='both', labelsize=14)
    for spine in ['right']:
        ax.spines[spine].set_visible(False)

# Unified legend from subplot 0
handles, labels = axs[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=4, fontsize=13, frameon=True)

# Layout adjustments
plt.tight_layout(rect=[0, 0, 1, 0.88])
plt.savefig('Figures/GPUTDMh.pdf', dpi=300, bbox_inches='tight', format='pdf')
plt.show()
```
